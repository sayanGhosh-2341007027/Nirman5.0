```python
# staffmgmt/management/commands/train_model.py

import os
import django
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "core.settings")
django.setup()

import pandas as pd
import joblib
from django.core.management.base import BaseCommand
from django.conf import settings

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

DATASET_PATH = os.path.join(settings.BASE_DIR, "core", "dataset", "understaffing_dataset.csv")
MODEL_PATH = os.path.join(settings.BASE_DIR, "core", "savedModels", "model.joblib")

class Command(BaseCommand):
    help = "Train ML model from CSV dataset"

    def handle(self, *args, **options):

        df = pd.read_csv(DATASET_PATH)

        # FEATURES + TARGET
        X = df[["scheduled_shifts", "emergencies_count", "patient_count"]]
        y = df["understaffed"]

        # ML PIPELINE
        pipeline = Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler()),
            ("clf", RandomForestClassifier(n_estimators=200, random_state=42))
        ])

        pipeline.fit(X, y)

        # Save Model
        joblib.dump(pipeline, MODEL_PATH)

        self.stdout.write(self.style.SUCCESS(f"Model saved to {MODEL_PATH}"))
```

```python
model = joblib.load("model.joblib")
prediction = model.predict(df)
```
