```python
from fastapi import FastAPI, Request
from fastapi.responses import HTMLResponse
from fastapi.staticfiles import StaticFiles
from fastapi.templating import Jinja2Templates

# Create FastAPI app
app = FastAPI()

# Tell FastAPI where templates and static files are
templates = Jinja2Templates(directory="templates")
app.mount("/static", StaticFiles(directory="static"), name="static")


# ---------- ROUTES ----------

# Home route -> return HTML page
@app.get("/", response_class=HTMLResponse)
async def home(request: Request):
    # "request" is required by Jinja2Templates
    return templates.TemplateResponse(
        "index.html",
        {
            "request": request,
            "title": "FastAPI + HTML + CSS + JS",
            "message": "Hello from FastAPI backend!"
        }
    )


# Simple API route for JS to call
@app.get("/api/hello")
async def api_hello():
    return {"msg": "This is data from FastAPI /api/hello route"}
```
